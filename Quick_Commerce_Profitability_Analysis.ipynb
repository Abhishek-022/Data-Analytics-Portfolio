```python
print("\n" + "=" * 80)
print("UNIT ECONOMICS BREAKDOWN (PER ORDER)")
print("Current State: 2024")
print("=" * 80)

unit_economics = {
    'Component': [
        'Average Order Value (AOV)',
        '- Product Cost (70%)',
        '= Gross Margin (30%)',
        '- Last Mile Delivery',
        '- Dark Store & Warehousing',
        '- Packaging & Wastage',
        '- Platform & Tech',
        '- Customer Support',
        '- Payment Gateway',
        '- Platform Fees',
        '= Contribution Margin 1 (CM1)',
        '- CAC (amortized)',
        '= Contribution Margin 3 (CM3)'
    ],
    'Amount_Rs': [485, -340, 145, -55, -39, -12, -8, -4, -10, -12, -15, -38, -53],
    'Percent_of_AOV': [100, -70, 30, -11.3, -8.0, -2.5, -1.6, -0.8, -2.1, -2.5, -3.1, -7.8, -10.9]
}

df_unit_econ = pd.DataFrame(unit_economics)

print("\n")
print(df_unit_econ.to_string(index=False))

print("\nâš ï¸  CRITICAL FINDING:")
print(f"  CM1 (before CAC): Rs -15 per order")
print(f"  CM3 (after CAC): Rs -53 per order")
print(f"  Status: BURNING CASH on every delivery")
```

```python
def calculate_unit_economics(aov, gross_margin_pct=0.30, delivery_cost=55, 
                             dark_store=39, packaging=12, platform=8,
                             support=4, payment_rate=0.02, fees=12, 
                             cac=300, avg_orders=8):
    """Custom unit economics calculator for quick commerce."""

    revenue = aov
    product_cost = aov * (1 - gross_margin_pct)
    gross_margin = aov * gross_margin_pct

    payment_cost = aov * payment_rate
    total_variable = (delivery_cost + dark_store + packaging + 
                     platform + support + payment_cost + fees)

    cm1 = gross_margin - total_variable
    cac_per_order = cac / avg_orders
    cm3 = cm1 - cac_per_order

    return {
        'aov': aov,
        'gross_margin': gross_margin,
        'delivery_cost': delivery_cost,
        'total_variable': total_variable,
        'cm1': cm1,
        'cm1_pct': (cm1/aov)*100,
        'cac_per_order': cac_per_order,
        'cm3': cm3,
        'cm3_pct': (cm3/aov)*100,
        'profitable': cm3 > 0
    }

# Test current state
current = calculate_unit_economics(485, delivery_cost=55, cac=300)

print("\n" + "=" * 80)
print("UNIT ECONOMICS CALCULATOR - CURRENT STATE")
print("=" * 80)

print(f"\nInputs:")
print(f"  AOV: Rs {current['aov']}")
print(f"  Delivery cost: Rs {current['delivery_cost']}")

print(f"\nOutputs:")
print(f"  CM1: Rs {current['cm1']:.0f} ({current['cm1_pct']:.1f}% of AOV)")
print(f"  CM3: Rs {current['cm3']:.0f} ({current['cm3_pct']:.1f}% of AOV)")

if not current['profitable']:
    print(f"\n  âš ï¸  UNPROFITABLE (burning Rs {abs(current['cm3']):.0f}/order)")
```

```python
print("\n" + "=" * 80)
print("BREAK-EVEN ANALYSIS")
print("=" * 80)

aov_range = np.arange(400, 700, 20)
results = []

for aov in aov_range:
    metrics = calculate_unit_economics(aov, delivery_cost=55, cac=300)
    results.append({
        'AOV': aov,
        'CM1': metrics['cm1'],
        'CM3': metrics['cm3'],
        'CM1_Positive': metrics['cm1'] > 0,
        'CM3_Positive': metrics['cm3'] > 0
    })

df_breakeven = pd.DataFrame(results)

cm1_breakeven = df_breakeven[df_breakeven['CM1_Positive']]['AOV'].min()
cm3_breakeven = df_breakeven[df_breakeven['CM3_Positive']]['AOV'].min()

print(f"\nBreak-even AOV (CM1): Rs {cm1_breakeven}")
print(f"  Current: Rs 485")
print(f"  Gap: Rs {cm1_breakeven - 485} ({((cm1_breakeven/485-1)*100):.1f}% increase)")

print(f"\nBreak-even AOV (CM3): Rs {cm3_breakeven}")
print(f"  Current: Rs 485")
print(f"  Gap: Rs {cm3_breakeven - 485} ({((cm3_breakeven/485-1)*100):.1f}% increase)")

print("\nðŸ’¡ PRIMARY LEVER: Increase AOV from Rs 485 to Rs 580-600")
```

```python
print("\n" + "=" * 80)
print("PROFITABILITY SCENARIOS (2028)")
print("=" * 80)

scenarios = {
    'Scenario': ['Conservative', 'Base Case', 'Optimistic'],
    'AOV': [550, 620, 720],
    'Delivery_Cost': [45, 38, 32],
    'Gross_Margin_%': [31, 32, 34],
    'CAC': [280, 240, 200],
    'Avg_Orders': [10, 12, 15]
}

df_scenarios = pd.DataFrame(scenarios)

results = []

for idx, row in df_scenarios.iterrows():
    metrics = calculate_unit_economics(
        aov=row['AOV'],
        gross_margin_pct=row['Gross_Margin_%']/100,
        delivery_cost=row['Delivery_Cost'],
        cac=row['CAC'],
        avg_orders=row['Avg_Orders']
    )

    results.append({
        'Scenario': row['Scenario'],
        'AOV': row['AOV'],
        'CM1': round(metrics['cm1'], 0),
        'CM3': round(metrics['cm3'], 0),
        'Status': 'âœ… Profitable' if metrics['cm3'] > 0 else 'âš ï¸ Loss',
        'Timeline': 'Q4 2027' if idx==0 else ('Q4 2026' if idx==1 else 'Q2 2026')
    })

df_results = pd.DataFrame(results)

print("\nScenario Outcomes:")
print(df_results.to_string(index=False))
```

```python
print("\n" + "=" * 80)
print("CITY-WISE PROFITABILITY ANALYSIS")
print("=" * 80)

city_data = {
    'City': ['Mumbai', 'Delhi NCR', 'Bangalore', 'Pune', 'Hyderabad'],
    'Rent_Lakhs': [1.20, 1.00, 0.85, 0.65, 0.70],
    'Daily_Orders': [400, 380, 420, 280, 260],
    'AOV': [520, 510, 500, 480, 470],
    'Density': ['High', 'High', 'High', 'Medium', 'Medium']
}

df_cities = pd.DataFrame(city_data)

city_profit = []

for idx, row in df_cities.iterrows():
    monthly_orders = row['Daily_Orders'] * 30

    metrics = calculate_unit_economics(aov=row['AOV'], delivery_cost=50, cac=280)

    fixed_costs = (row['Rent_Lakhs'] * 100000 + 213000 + 45000 + 20000 + 5000)

    store_cm1 = metrics['cm1'] * monthly_orders
    store_profit = store_cm1 - fixed_costs
    profit_per_order = store_profit / monthly_orders

    city_profit.append({
        'City': row['City'],
        'Daily_Orders': row['Daily_Orders'],
        'AOV': row['AOV'],
        'Store_Profit_Lakhs': round(store_profit/100000, 1),
        'Status': 'âœ…' if store_profit > 0 else 'âš ï¸',
        'Breakeven_Orders': round(fixed_costs / abs(metrics['cm1']))
    })

df_city_profit = pd.DataFrame(city_profit)

print("\n")
print(df_city_profit.to_string(index=False))

pune_data = df_city_profit[df_city_profit['City']=='Pune'].iloc[0]
print(f"\nPune (My Home City):")
print(f"  Current: {pune_data['Daily_Orders']} orders/day")
print(f"  Breakeven: {pune_data['Breakeven_Orders']} orders/day")
print(f"  Gap: {pune_data['Breakeven_Orders'] - pune_data['Daily_Orders']} more orders needed")
```

```python
print("\n" + "=" * 80)
print("MONTE CARLO SIMULATION: 10,000 SCENARIOS")
print("=" * 80)

sim_results = []
np.random.seed(42)

for sim in range(10000):
    sim_aov = np.random.normal(485, 50)
    sim_aov = np.clip(sim_aov, 350, 650)

    sim_delivery = np.random.normal(50, 5)
    sim_delivery = np.clip(sim_delivery, 35, 70)

    sim_cac = np.random.normal(280, 40)
    sim_cac = np.clip(sim_cac, 180, 400)

    sim_orders = np.random.normal(10, 2)
    sim_orders = np.clip(sim_orders, 5, 18)

    metrics = calculate_unit_economics(
        aov=sim_aov,
        delivery_cost=sim_delivery,
        cac=sim_cac,
        avg_orders=sim_orders
    )

    sim_results.append({
        'CM3': metrics['cm3'],
        'Profitable': metrics['cm3'] > 0
    })

df_sim = pd.DataFrame(sim_results)

print(f"\nâœ… Simulation complete!")
print(f"\nResults:")
print(f"  Profitable scenarios: {df_sim['Profitable'].sum():,} ({df_sim['Profitable'].mean()*100:.1f}%)")
print(f"  Mean CM3: Rs {df_sim['CM3'].mean():.2f}")
print(f"  Median: Rs {df_sim['CM3'].median():.2f}")
print(f"  10th percentile: Rs {df_sim['CM3'].quantile(0.1):.2f}")
print(f"  90th percentile: Rs {df_sim['CM3'].quantile(0.9):.2f}")
```
